```python
import numpy as np
import plotly.graph_objects as go
import matplotlib.cm as cm
import matplotlib.colors as colors
from pathlib import Path
import pickle
import xarray as xr
from tqdm.notebook import tqdm
```

```python
# read data
PROJECT_ROOT = Path.cwd().parent
pkl_file=PROJECT_ROOT/"result/run_approach.pkl"

with open(pkl_file, "rb") as f:
    obj = pickle.load(f, encoding="latin1")

    # 提取所需变量并转为 numpy array
    density = np.asarray(obj['densities'])
    reference_levels = np.asarray(obj['reference_levels'])
    score_refden = np.asarray(obj['scores_refden'])
    score_regul = np.asarray(obj['scores_regul'])
    regul = np.asarray(obj['regul_params'])
    regul_residual = np.asarray(obj['regul_residuals'])
    refden_residual = np.asarray(obj['refden_residuals'])

    refden_moho_grid = obj['best_solutions_refden_moho_grid']
    refden_predict_grid = obj['best_solutions_refden_predict_grid']
    observe = obj['observe']
    lon = obj['lon']
    lat = obj['lat']
    lon_sub = obj['lon_sub']
    lat_sub = obj['lat_sub']
    
nlat_moho, nlon_moho = refden_moho_grid.shape
lon_moho = np.linspace(lon.min(), lon.max(), nlon_moho)
lat_moho = np.linspace(lat.min(), lat.max(), nlat_moho)

moho_grid = xr.DataArray(
    refden_moho_grid,  # 如果图东西颠倒或拉伸，改为 refden_moho_grid.T
    dims=("lat", "lon"),
    coords={"lat": lat_moho, "lon": lon_moho}
)
```

```python
moho_grid.values=moho_grid.values*1e-3+40
```

```python
# ==========================
# 基本参数
# ==========================

R0 = 173 - 40  # 月球半径-平均壳厚 km

DLON = float(moho_grid.lon[1] - moho_grid.lon[0])
DLAT = float(moho_grid.lat[1] - moho_grid.lat[0])

# ==========================
# 球坐标 -> 笛卡尔
# ==========================

def sph2cart(lon_deg, lat_deg, r):

    lon = np.deg2rad(lon_deg)
    lat = np.deg2rad(lat_deg)

    x = r * np.cos(lat) * np.cos(lon)
    y = r * np.cos(lat) * np.sin(lon)
    z = r * np.sin(lat)

    return np.array([x, y, z])


# ==========================
# 球面棱柱体
# ==========================

def spherical_prism(lon1, lon2, lat1, lat2, r1, r2):

    corners = [
        (lon1, lat1),
        (lon2, lat1),
        (lon2, lat2),
        (lon1, lat2)
    ]

    inner = [sph2cart(lon, lat, r1) for lon, lat in corners]
    outer = [sph2cart(lon, lat, r2) for lon, lat in corners]

    verts = np.array(inner + outer)

    faces = []

    # bottom
    faces += [[0,1,2],[0,2,3]]

    # top
    faces += [[4,6,5],[4,7,6]]

    # sides
    sides = [
        [0,1,5,4],
        [1,2,6,5],
        [2,3,7,6],
        [3,0,4,7]
    ]

    for a,b,c,d in sides:
        faces += [[a,b,c],[a,c,d]]

    return verts, np.array(faces)


# ==========================
# colormap
# ==========================

cmap = cm.get_cmap("seismic")

dhs = moho_grid.values.flatten()

norm = colors.Normalize(
    vmin=np.min(dhs),
    vmax=np.max(dhs)
)


# ==========================
# Plotly 图
# ==========================

fig = go.Figure()

# MOS 135E 160E 15N 35N

LONMIN,LONMAX=135,160
LATMIN,LATMAX=15,35

moho_grid=moho_grid.sel(lon=slice(LONMIN,LONMAX),lat=slice(LATMIN,LATMAX))

lons = moho_grid.lon.values
lats = moho_grid.lat.values

length=np.size(lons)*np.size(lats)

index=0
for lon in tqdm(lons,desc="Outer"):
    for lat in tqdm(lats,desc="Inner",leave=False):
        

        dh = moho_grid.sel(lon=lon, lat=lat).item()

        r1 = R0
        r2 = R0 + dh

        r_inner = min(r1, r2)
        r_outer = max(r1, r2)

        verts, faces = spherical_prism(
            lon,
            lon + DLON,
            lat,
            lat + DLAT,
            r_inner,
            r_outer
        )

        x, y, z = verts.T
        i, j, k = faces.T

        # 颜色
        color = cmap(norm(dh))
        color = f"rgb({int(color[0]*255)},{int(color[1]*255)},{int(color[2]*255)})"

        # 棱柱
        fig.add_trace(go.Mesh3d(
            x=x,
            y=y,
            z=z,
            i=i,
            j=j,
            k=k,
            color=color,
            opacity=0.9,
            flatshading=True,
            showlegend=False
        ))

        # ======================
        # edges
        # ======================

        edges = [
            (0,1),(1,2),(2,3),(3,0),
            (4,5),(5,6),(6,7),(7,4),
            (0,4),(1,5),(2,6),(3,7)
        ]

        for a,b in edges:

            fig.add_trace(go.Scatter3d(
                x=[x[a],x[b]],
                y=[y[a],y[b]],
                z=[z[a],z[b]],
                mode="lines",
                line=dict(color="black", width=1),
                showlegend=False,
                hoverinfo="skip"
            ))
            
        index=index+1

# ==========================
# layout
# ==========================

fig.update_layout(

    title="Moho Spherical Prism Model",

    scene=dict(

        xaxis_title="X (km)",
        yaxis_title="Y (km)",
        zaxis_title="Z (km)",

        aspectmode="data"
    )
)

fig.show()
```
